```python
# Define the ticker symbol
tickerSymbol = 'AAPL'

# Get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get the historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2016-02-19', end='2023-06-30')

# See the data
tickerDf = tickerDf.rename_axis('date')

tickerDf = tickerDf.drop(columns= ['Dividends','Stock Splits'])

tickerDf.index = tickerDf.index.strftime("%Y-%m-%d")

tickerDf.tail()
```

```python
# Set your API token and desired parameters
api_token = "649e199732fe84.28932233"
tickers = "aapl"
start_date = "2007-06-01"
end_date = "2023-06-30"

# Construct the URL with the parameters
url = f"https://eodhistoricaldata.com/api/sentiments?s={tickers}&from={start_date}&to={end_date}&api_token={api_token}"

# Make the GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print("Error:", response.status_code)

```

```python
ticker = list(data.keys())[0]
# Get the data list for the ticker
data_list = data[ticker]
# Create DataFrame from the data list
df = pd.DataFrame(data_list).set_index("date")

df = df.drop(columns=['count'])
# Print the DataFrame
df.head()

```

```python
combined_data = pd.concat([df, tickerDf], axis= 1)
combined_data = combined_data.dropna()

combined_data
```

```python
combined_data.index = pd.to_datetime(combined_data.index)

combined_data = combined_data.sort_index()

combined_data
```

In the combined_data DF normalized represent the news sentiment for the security for that particular day rangin from 0 (bad sentiment) 0.1 really good sentiment

```python

# Assuming you have your input features in X and target variable (prices) in y
X = combined_data.drop(columns=['Close'])
y= combined_data['Close'].copy()

X
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
from pandas.tseries.offsets import DateOffset


training_begin = X.index.min()
training_end = X.index.min() + DateOffset(months=72)

# Generate the X_train and y_train DataFrames
X_train = X.loc[training_begin:training_end]
y_train = y.loc[training_begin:training_end]

# Display sample data
training_begin

X_train
```

```python
# Generate the X_test and y_test DataFrames
X_test = X.loc[training_end:]
y_test = y.loc[training_end:]

# Display sample data
X_test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression

# Initialize the input features scaler
scaler_X = MinMaxScaler(feature_range=(0, 1))
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

# Initialize the target variable scaler
scaler_y = MinMaxScaler(feature_range=(0, 1))
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1))

```

**Model 1: Linear Regression**

This model is using news sentiment as a parameter

```python
# Train the Linear Regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
predictions_scaled = model.predict(X_test_scaled)

# Transform the predictions back to the original scale
predictions = scaler_y.inverse_transform(predictions_scaled)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R2 Score:", r2)

feature_names = X_train.columns

# Get the weights (coefficients) assigned to each feature
weights = model.coef_ * 100

# Print the feature names with their corresponding weights
for feature, weight in zip(feature_names, weights[0]):
    print(f"{feature}: {weight}")
```
